```python
import os 
import nltk
import nltk.corpus
import pandas as pd
from nltk.corpus import stopwords
import nltk
import numpy as np
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
def docs(*doc_path):
    doc_list=[]
    for p in doc_path:
        data=open(p)
        y=data.read()
        punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
        for ele in y:
            if ele in punc:
                y=y.replace(ele,"")
        y=y.lower()
        
        tokens= [t for t in y.split()]
        nltk.download('stopwords')
        sr= stopwords.words("english")
        clean_tokens=tokens[:]
        for token in tokens:
            if token in sr:
                clean_tokens.remove(token)
    
        doc_list.append(clean_tokens)
    return doc_list
```
