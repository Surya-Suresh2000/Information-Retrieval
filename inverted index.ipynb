```python
import os
import nltk
import nltk.corpus
import pandas as pd
import string
import numpy as np
from nltk.corpus import stopwords
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
def docs(*doc_path):
    doc_list=[]
    for p in doc_path:
        data=open(p)
        y=data.read()
        punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
        for ele in y:
            if ele in punc:
                y=y.replace(ele,"")
        y=y.lower()
        
        tokens= [t for t in y.split()]
        nltk.download('stopwords')
        sr= stopwords.words("english")
        clean_tokens=tokens[:]
        for token in tokens:
            if token in sr:
                clean_tokens.remove(token)
    
        doc_list.append(clean_tokens)
    return doc_list
```

```python
doc_list=docs(r"D:\doc_txt1.txt",r"D:\doc_txt2.txt",r"D:\doc_txt3.txt")
```

```python
import itertools
m=list(itertools.chain.from_iterable(doc_list))
unique = []
for w in m:
    if w not in unique:
        unique.append(w)
print(unique)
```

```python
from collections import defaultdict
docs=defaultdict(list)
for word in unique:
    if word!=" " or word!='':
        if word in doc_list[0]:
            docs[word].append('1')
        else :
            pass
        if word in doc_list[1]:
            docs[word].append('2')
        else:
            pass
        if word in doc_list[2]:
            docs[word].append('3')
        else:
            pass
    else:
        continue
```
